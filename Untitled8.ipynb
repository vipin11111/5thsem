```python
import pandas as pd
df = pd.read_excel('dataset/SaleData.xlsx')
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df1=df.groupby(df['Region']).sum()[['Sale_amt']]
df1
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_new = df.groupby('Region').sum()['Sale_amt']
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0.5,0.5,1,1])

#plotting the lines
ax.bar(df_new.index,df_new)
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_new = df.groupby('Item').sum()['Units']
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('item wise units sold')
ax.set_xlabel('items')
ax.set_ylabel('units_sold')
#plotting the lines
ax.bar(df_new.index,df_new)
```
