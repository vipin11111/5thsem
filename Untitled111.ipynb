## 1.Reading the data


```python
import pandas as pd
df1 = pd.read_csv("dataset/titanic.csv")
df1
```

## 2. Checking if there are missing values

```python
df1.isnull().sum()
```

## 3.Filling missing values with 0

```python
new_df1 = df1.fillna(0)
new_df1
```

## 4. Filling NaN values with forward fill value

```python
new_df2 = df1.fillna(method="ffill")
new_df2
```

## 5. Setting forward fill limit to 1

```python
new_df3 = df1.fillna(method="ffill",limit=1)
new_df3
```

## 6. Filling NaN values in Backward Direction

```python
new_df4 = df1.fillna(method="bfill")
new_df4
```

## 7. Interpolate of missing values

```python
new_df5 = df1.interpolate() 
new_df5
```

## 8. Dropna()

```python
new_df6 = df.dropna()
new_df6
```

## 9. Deleting the rows having all NaN values

```python
new_df7 = df.dropna(how='all')
new_df7
```
