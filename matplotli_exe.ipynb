```python
import pandas as pd
df=pd.read_excel('dataset/SaleData.xlsx')
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

## using matplotlib bar chart display total amount of sales region wise

```python
df.groupby('Region').sum()[['Sale_amt']]
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_new = df.groupby('Region').sum()['Sale_amt']
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('sales amount per region')
ax.set_xlabel('(region)')
ax.set_ylabel('(sales amount)')
#plotting the lines
ax.bar(df_new.index,df_new)
```

## using matploptlib bar chart display the item wise unit sold

```python
df.groupby('Item').sum()[['Units']]
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_new = df.groupby('Item').sum()['Units']
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('item wise units sold')
ax.set_xlabel('(items)')
ax.set_ylabel('(units_sold)')
#plotting the lines
ax.bar(df_new.index,df_new)
```
